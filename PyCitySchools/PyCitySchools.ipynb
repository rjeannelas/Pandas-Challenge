```python
# Dependencies and setup
import pandas as pd

# File to load (remember to change these)
school_data_to_load = "Downloads\schools_complete.csv"
student_data_to_load = "Downloads\students_complete.csv"

# Read school and student data file and store into pandas dataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
# Calculate the total number of schools
total_unique_schools = school_data_complete["school_name"].nunique()
total_unique_schools
```

```python
# Calculate the total number of students
total_students_number = school_data_complete["Student ID"].count()
total_students_number
```

```python
# Calculate the total budget
total_budget = school_data_complete.drop_duplicates(subset="school_name")
TBudget = total_budget["budget"].sum()
TBudget
```

```python
# Calculate the average math score
average_math_score = school_data_complete["math_score"].mean()
average_math_score
```

```python
# Calculate the average reading score
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
students_passing_math = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
percent_passing_math = (number_students_passing_math / total_students_number)*100
percent_passing_math
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)
students_passing_reading = school_data_complete[school_data_complete["reading_score"] >=70]
number_students_passing_reading = students_passing_reading["Student ID"].count()
percent_passing_reading = (number_students_passing_reading / total_students_number)*100
percent_passing_reading
```

```python
# Calculate the percentage of students who passed math and reading (% overall passing)
overall_passing = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)].count()["student_name"]
overall_percentage_rate = (overall_passing / total_students_number)*100
overall_percentage_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a dataframe to hold the above results
district_summary = pd.DataFrame(
{
    "Total Schools": [total_school_number],
    "Total Students": f"{total_students_number:,}",
    "Total Budget": f"${TBudget:,.2f}",
    "Average Math Score": f"{average_math_score:.6f}",
    "Average Reading Score": f"{average_reading_score:.5f}",
    "% Passing Math": f"{percent_passing_math:.6f}",
    "% Passing Reading": f"{percent_passing_reading:.6f}",
    "% Overall Passing": f"{overall_passing: .6f}"
}, index=[0])

# Formatting
district_summary
```

```python
# Group by school name
school_name = school_data_complete.set_index("school_name").groupby(["school_name"])

# School types by school name
school_type = school_data.set_index(["school_name"])["type"]
school_type_df = school_type.to_frame("School Type")
school_type_df
```

```python
# Calculate total students
total_student = school_name["Student ID"].count()
total_student_df = total_student.to_frame("Total Students")
total_student_df
```

```python
# Total school budget
total_school_budget = school_data.set_index("school_name")["budget"]
total_school_budget_df = total_school_budget.to_frame("Total School Budget")
total_school_budget_df
```

```python
# Per student budget
budget_per_student = (school_data.set_index("school_name")["budget"]/school_data.set_index("school_name")["size"])
budget_per_student_df = budget_per_student.to_frame("Budget Per Student")
budget_per_student_df
```

```python
# Average Math Score
average_math_score = school_name["math_score"].mean()
average_math_score_df = average_math_score.to_frame("Per Student Budget")
average_math_score_df
```

```python
# Average Reading Score
average_reading_score = school_name["reading_score"].mean()
average_reading_score_df = average_reading_score.to_frame("Average Math Score")
average_reading_score_df
```

```python
# Percentage of students with a passing math score (70 or greater)
pass_math_percent = school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name")["Student ID"].count()/total_student*100
pass_math_percent_df = pass_math_percent.to_frame("Percentage Passing Math")
pass_math_percent_df
```

```python
# Percentage of students with a passing reading score (70 or greater)
pass_reading_percent = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name")["Student ID"].count()/total_student*100
pass_reading_percent_df = pass_reading_percent.to_frame("Percentage Passing Reading")
pass_reading_percent_df
```

```python
# Percentage Overall Passing (the percentage of students who passed math AND reading)

overall_pass =  school_data_complete[(school_data_complete["reading_score"] >= 70) &\
                                     (school_data_complete["math_score"]>= 70)].groupby\
                                     (["school_name"])["student_name"].size()
overall_percentage_rate = overall_pass / total_student * 100
overall_percentage_rate_df = overall_percentage_rate.to_frame("Percentage Overall Passing")
overall_percentage_rate_df
```

```python
# Combine into a a dataframe called School Summary
school_summary = pd.DataFrame({
    "School Type": school_type,
    "Total Students": total_student,
    "Total Budget": total_school_budget,
    "Per Student Budget": budget_per_student,
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    '% Passing Math': percent_passing_math,
    '% Passing Reading': percent_passing_reading,
    "% Overall Pass": overall_percentage_rate
})

school_summary = school_summary[['School Type',
                        'Total Students',
                        'Total Budget',
                        'Per Student Budget',
                        'Average Math Score',
                        'Average Reading Score',
                        '% Passing Math',
                        '% Passing Reading',
                        '% Overall Pass']]

# Formatting
school_summary.style.format({'Total Students': '{:}',
                             "Total Budget": "${:,.2f}",
                             "Per Student Budget": "${:.2f}",
                             'Average Math Score': "{:6f}",
                             'Average Reading Score': "{:6f}",
                             "% Passing Math": "{:6f}",
                             "% Passing Reading": "{:6f}",
                             "% Overall Pass": "{:3f}"
                            })
```

```python
# Highest-Performing Schools (by % Overall Passing)
top_perform = school_summary.sort_values(by="% Overall Pass", ascending = False)
top_perform.head()
```

```python
# Lowest-Performing Schools (by % Overall Passing)

bottom_schools = school_summary.sort_values(by="% Overall Pass", ascending=True)
bottom_schools.head()
```

```python
# Math Scores by Grade
ninth_math = student_data.loc[student_data['grade'] == '9th'].groupby('school_name')["math_score"].mean()
tenth_math = student_data.loc[student_data['grade'] == '10th'].groupby('school_name')["math_score"].mean()
eleventh_math = student_data.loc[student_data['grade'] == '11th'].groupby('school_name')["math_score"].mean()
twelfth_math = student_data.loc[student_data['grade'] == '12th'].groupby('school_name')["math_score"].mean()

math_scores = pd.DataFrame({
        "9th": ninth_math,
        "10th": tenth_math,
        "11th": eleventh_math,
        "12th": twelfth_math
})
math_scores = math_scores[['9th', '10th', '11th', '12th']]
math_scores.index.name = "School Name"

# Formatting
math_scores.style.format({'9th': '{:.6f}', 
                          "10th": '{:.6f}', 
                          "11th": "{:.6f}", 
                          "12th": "{:.6f}"})
```

```python
# Reading Scores by Grade
ninth_reading = student_data.loc[student_data['grade'] == '9th'].groupby('school_name')["reading_score"].mean()
tenth_reading = student_data.loc[student_data['grade'] == '10th'].groupby('school_name')["reading_score"].mean()
eleventh_reading = student_data.loc[student_data['grade'] == '11th'].groupby('school_name')["reading_score"].mean()
twelfth_reading = student_data.loc[student_data['grade'] == '12th'].groupby('school_name')["reading_score"].mean()

reading_scores = pd.DataFrame({
        "9th": ninth_reading,
        "10th": tenth_reading,
        "11th": eleventh_reading,
        "12th": twelfth_reading
})
reading_scores = reading_scores[['9th', '10th', '11th', '12th']]
reading_scores.index.name = "School Name"

# Formatting
reading_scores.style.format({'9th': '{:.6f}', 
                             "10th": '{:.6f}', 
                             "11th": "{:.6f}", 
                             "12th": "{:.6f}"})
```

```python
# Scores by School Spending

# Create four bins with reasonable cutoff values to group school spending
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Categorize the spendings 
school_summary["Spending Ranges (Per Student)"] = pd.cut(budget_per_student,spending_bins,labels = labels)

# Scores Based on Spending Ranges
average_math_score_spending = round(school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"],2)
average_reading_score_spending = round(school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"],2)
students_passing_math_spending = round(school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"],2)
students_passing_reading_spending = round(school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"],2)
overall_pass_rate_spending = round(school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Pass"],2)


spending_summary = pd.DataFrame({"Average Math Score":avg_math_score_spending,
                                     "Average Reading Score":avg_reading_score_spending,
                                     "% Passing Math":students_passing_math_spending,
                                     "%Passing Reading":students_passing_reading_spending ,
                                     "% Overall Passing":overall_pass_rate_spending})
spending_summary
```

```python
# Scores by School Size
# Create a dataframe that breaksdown school performance based on school size

size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
school_summary = school_summary.copy()
school_summary["School Size"] = pd.cut(school_summary['Total Students'], bins=size_bins, labels=labels)

size_summary = school_summary.groupby("School Size")
size_summary = size_summary[["Average Math Score", "Average Reading Score", "% Passing Math", 
                             "% Passing Reading", "% Overall Pass"]]
size_summary.mean()
```
